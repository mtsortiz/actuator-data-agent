```python
import pdfplumber
from dotenv import load_dotenv
from langchain_chroma.vectorstores import Chroma
from langchain_openai import OpenAIEmbeddings
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
from pathlib import Path
```

```python
load_dotenv()
```

```python
file_path_summary = "D:\\Development\\intelligent-actuator-data-agent\\data\\raw\\series_76_summary.pdf"
```

```python
#Extract
text = ""
with pdfplumber.open(file_path_summary) as pdf:
    for page in pdf.pages:
        text += page.extract_text() + "\n"
```

```python
#Chunking
text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=500, 
        chunk_overlap=50)
docs = text_splitter.create_documents([text])
```

```python
embeddings = GoogleGenerativeAIEmbeddings(model="gemini-embedding-2-preview")
#embeddings = OpenAIEmbeddings(model="text-embedding-3-small")
```

```python
root_dir = Path.cwd().parent
chroma_path = root_dir / "data" / "processed" / "chroma_db"

Chroma.from_documents(docs, embeddings, persist_directory=str(chroma_path))
```
